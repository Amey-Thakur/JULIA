## Make some noisy data

```julia
xs = 0:0.05:5
ys = 1 .+ 0.5*(sqrt.(xs) .+ 0.8 * rand(length(xs))).^2;
```

```julia
using Plots
```

```julia
scatter(xs, ys, ylim=(0, 6))
```

## How can we fit a straight line to this data?

```julia
# ax + b

line(x, a, b) = a * x + b
```

```julia
a = 0.1
b = -0.3

scatter(xs, ys, ylim=(0, 6))
plot!(0:0.01:6, x -> line(x, a, b))
```

```julia
using Interact
```

```julia
@manipulate for a in -1:0.01:1, b in -1:0.01:1

    scatter(xs, ys, ylim=(0, 6))
    plot!(0:0.01:6, x -> line(x, a, b))
    
end
```
